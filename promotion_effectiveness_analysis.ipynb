```python
sns.scatterplot(x='Discount_%', y='Profit', hue='Promotion', data=df, s=100)
plt.title("Discount vs Profit")
plt.show()
```

## Business Insights

1. Promotions increase sales quantity, but heavy discounts can reduce profit.
2. Moderate discounts (10-15%) provide the best balance between sales and profit.
3. Shampoo responds better to promotions than Soap in our dataset.
4. Store B has slightly better sales during promotions than Store A.

## Recommendations

1. Apply promotions selectively for products with high response.
2. Avoid very high discounts to protect profit margins.
3. Monitor store-wise performance to optimize promotions.
4. Regularly analyze promotion effectiveness for better decision-making.
